```python
import pandas as pd

#read csv and sort 'Duration' largest to smallest
df = pd.read_csv('files/covid19_tweets.csv').sort_values(by=['user_location'], ascending=False)

print("--Top 10 User Locations---------")
print(df.head(10))

print("--Bottom 10 User Locations------")
print(df.tail(10))


```

I used pandas to work with the file. I wanted to use selection methods (selecting top 10 and bottom 10 locations) since I wanted to check for any sort of recent locations that people have been tweeting about, and then checked the bottom 10 locations to check for any similar trends occurring at an earlier time. Pandas is very useful for this, as it is easy to filter by category in a CSV as such that is labeled by column.
